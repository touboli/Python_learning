```python
"""使用Python打开一个文本文件"""
f = open('test.txt','r',encoding = 'utf-8')  #打开文件，r表示以只读的方式打开文件
print(f.read())    #读取文件内容
f.close()  #关闭文件句柄
```

```python
f = open("test.txt","r",encoding = "utf-8")
print(f.readline())  #读取一行
print(f.read(5))
f.close()

"""read()方法会读取文件的全部内容，并返回一个字符串
。如果你在调用read()方法之后再调用readline()方法，
readline()方法将会返回一个空字符串，因为read()方法
已经将文件指针移动到了文件的末尾。
readline()方法会读取文件的一行内容，并返回一个字符串。
\每次调用readline()方法，文件指针都会移动到下一行，
所以下一次调用readline()方法时，它将会返回下一行的内容。"""
```

```python
"""写入内容到文本中"""
str_ = """南宋末年，金兵入侵，朝廷奸臣当道。全真派丘处机因杀死汉奸王道乾，被官兵追杀，逃到江南一个小村牛
家庄， 
与隐居在此处的爱国义土郭啸天和杨铁心一见如故，以“靖、康”为他俩即将出生的孩子命名，并留下一对
短剑作为信物。 
三人酒醉之后。杨妻包惜弱在外面忽然发现一个受伤的官兵，于是偷偷将其救下。 
哪知却为郭、杨两家带来无尽的灾难。原来，此人竟是金国六王子完颜洪烈，因为追杀丘处机被打伤，得包
惜弱相救后，竟然对她一见钟情， 
于是勾结南宋官员段天德对郭、杨迫害。郭啸天死亡、杨铁心不知去向，已怀身孕的郭夫人李萍被段天德抓
走，之后经过千辛万苦逃到大漠， 
生下儿子郭靖，而杨夫人包惜弱则被骗嫁给完颜洪烈，生子完颜康。丘处机得知两家被害的消息后，一路对
凶手进行追查， 
与江南七怪发生误会。双方不打不相识，定下赌约：分别寻找杨、郭的后人并传授他们武功，于十八年后在
醉仙楼比试高低。 """

f = open("test.txt","r+",encoding = "utf-8")  #以读写模式打开文件 ,r+表示读写模式,w表示写入模式
f.write(str_)
print(f.readline())
f.close()
```

```python
"""使用a模式写入文件"""
f = open("test.txt","a+",encoding = "utf-8")  #以追加读写模式打开文件
f.write("这是追加的内容")
print(f.readline())
f.close()
```

```python
f=open("test.txt","a+",encoding = "utf-8")
print(f.readline())    #读取一行，因为文件指针已经移动到了文件的末尾，所以读取的是空字符串
f.close()
```

```python
"""换个姿势读文件"""
f = open("test.txt","r",encoding = "utf-8")
line = f.readline() 
lines = f.readlines() 
print(line) 
print(lines) 
```

```python
#写入文件不关闭的后果 
str_ = """ -- 射雕英雄传/金庸 
""" 
f = open("text_new.txt", "a", encoding="utf8") 
f.write(str_) 
# 此处不适用close关闭文件 
#读取这个没保存的文件： 
f_new = open("text_new.txt", "r", encoding="utf8") 
f_new.read() 
# 输出为空 
```

```python
"""使用flush()方法强制将缓冲区的数据写入文件"""
str_ = """ 
-- 射雕英雄传/金庸
"""
f = open("test_new.txt", "a", encoding="utf8")
f.write(str_)
f.flush()
f_new = open("test_new.txt", "r", encoding="utf8")
print(f_new.read())
```

```python
"""上下文管理器"""
with open("test.txt","a",encoding = "utf-8") as f:
    f.write("这小说真好看个鬼")

with open("test.txt","r",encoding = "utf-8") as f:
    print(f.read())
```
